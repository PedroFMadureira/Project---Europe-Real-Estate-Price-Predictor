```python
# Getting unique year values
years = []
for year in range(2010, 2023):
    for quarter in ['Q1', 'Q2', 'Q3', 'Q4']:
        years.append(f"{year}-{quarter}")

# Dataframe
years_df = pd.DataFrame({'year': years})
```

```python
cities_lst = list(cities_unique_df['city'])
years_lst = list(years_df['year'])

combinations = list(itertools.product(cities_lst, years_lst))

# Dataframe
cities_years_df = pd.DataFrame(combinations, columns=['city', 'year'])
cities_years_df = pd.merge(cities_years_df, cities_unique_df, on=['city'], how='outer')
cities_years_df = cities_years_df[['city_id','city','country','year']]
cities_years_df
```

```python
# Function to create dates based on year
def create_dates(row):
    year, quarter = row['year'].split('-')
    if quarter == 'Q1':
        return datetime(int(year), 3, 31).date()
    elif quarter == 'Q2':
        return datetime(int(year), 6, 30).date()
    elif quarter == 'Q3':
        return datetime(int(year), 9, 30).date()
    elif quarter == 'Q4':
        return datetime(int(year), 12, 31).date()

# Applying the function
cities_years_df['date'] = cities_years_df.apply(create_dates, axis=1)
cities_years_df['date'] = pd.to_datetime(cities_years_df['date'])
cities_years_df
```

```python
date_obj = datetime(2010, 12, 31)
n = 12
unsupported_currencies = []
currency_years = []

for currency in eu_currency_list:
    for i in range(n):
        date = date_obj + pd.DateOffset(years=i)
        try:
            converted_amount = c.convert(currency, 'EUR', 1, date)
        except:
            unsupported_currencies.append(currency)
            currency_years.append(2010+i)

unsupported_currencies_unique = sorted(set(unsupported_currencies))
currency_years_unique = sorted(set(currency_years))
```

```python
# Creating a dataframe for the exchange rates

eu_currencies = eu_currency_list.copy()
start_year = 2010
end_year = 2022

exchange_rates = {}

for currency in eu_currencies:
    exchange_rates[currency] = []
    for year in range(start_year, end_year + 1):
        date = pd.to_datetime(f'{year}-12-31')
        try:
            rate = c.get_rate(currency, 'EUR', date)
            exchange_rates[currency].append(rate)
        except:
            exchange_rates[currency].append(None)

# Dataframe
eu_rates_df = pd.DataFrame(exchange_rates)
eu_rates_df['year'] = range(start_year, end_year + 1)
eu_rates_df = eu_rates_df[['year'] + eu_currencies]
eu_rates_df['year'] = eu_rates_df['year'].astype(str)
eu_rates_df.loc[eu_rates_df['year'].str.len() == 4, 'year'] += '-Q4'
eu_rates_df['date'] = eu_rates_df.apply(create_dates, axis=1)
eu_rates_df['date'] = pd.to_datetime(eu_rates_df['date'])

eu_rates_df
```

```python
nan_cols = eu_rates_df.columns[eu_rates_df.isnull().any()].tolist()
eu_rates_df[nan_cols+['date']]
```

```python
# Hardcode values for unsupported currencies
ALL = [0.0072,0.00713,0.00715,0.00713,0.00715,0.0073,0.00733,0.00748,0.00803,0.00816,0.00808,0.00825,0.00856]
BAM = [0.51129] * len(eu_rates_df)
BYN = [0.000005,0.000005,0.000005,0.000005,0.000005,0.000005,0.48068,0.42237,0.41545,0.4407,0.32697,0.34769,0.38285]
MDL = [0.06299,0.06322,0.06221,0.05635,0.0535,0.04669,0.04659,0.04868,0.05116,0.05223,0.04876,0.05029,0.04966]
MKD = [0.01623,0.01626,0.01626,0.01625,0.01621,0.01624,0.01627,0.01625,0.01626,0.01626,0.01621,0.01621,0.01621]
RSD = [0.00935,0.00967,0.00893,0.00877,0.0083,0.00827,0.00811,0.00837,0.00845,0.00851,0.0085,0.0085,0.00852]
UAH = [0.09443,0.09461,0.09628,0.09205,0.05351,0.03947,0.03676,0.03142,0.03125,0.03778,0.02949,0.03281,0.02638]
# rates gotten from url below and BYN assumed as 0.000005 until 2015 as per web search
# https://commission.europa.eu/funding-tenders/procedures-guidelines-tenders/information-contractors-and-beneficiaries/exchange-rate-inforeuro_en
```

```python
eu_rates_df2 = eu_rates_df.copy()
eu_rates_df2['ALL'] = ALL
eu_rates_df2['BAM'] = BAM
eu_rates_df2['BYN'] = BYN
eu_rates_df2['MDL'] = MDL
eu_rates_df2['MKD'] = MKD
eu_rates_df2['RSD'] = RSD
eu_rates_df2['UAH'] = UAH

eu_rates_df2
```

```python
# Function to convert amounts to EUR using the exchange rates df
def currency_converter(row):
    currency = row['currency']
    date = row['date']
    amount = row['amount']

    converted_amount = amount * eu_rates_df2[eu_rates_df2['date'] == date][currency].values[0]
    return converted_amount

# Applying the function
hist_cities_3 = hist_cities_2.copy()
hist_cities_3['date'] = pd.to_datetime(hist_cities_3['date'])
hist_cities_3['amount_EUR'] = hist_cities_3.apply(currency_converter, axis=1)
hist_cities_3
```
